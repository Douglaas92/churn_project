# Projeto de Análise Exploratória de Churn

## 1. Definição e Escopo do Problema

### 1.1 Origem dos Dados:
- Telco (empresa fictícia de Telecomunicações - dataset disponibilizado no site do Kaggle.com)
- Link para o dataset: [Telco - Kaggle]([https://www.kaggle.com/datasets/blastchar/telco-customer-churn])

### 1.2 Contexto de Negócio:
Sabemos que o setor de telecomunicações segue em rápida expansão e transformação, principalmente em decorrência à digitalização, automação e avanços tecnológicos, como a expansão da rede 5G, por exemplo. Desta forma, as empresas enfrentam concorrência crescente de provedores menores, porém, gigantes em tecnologia, oferencendo serviços e soluções alternativas em grande margem. Diante deste cenário, o Churn (cancelamento de clientes) é um dos maiores desafios enfrentados pelas organizações do ramo Telecom, uma vez que há uma concorrência expressiva de preços agressivos, problemas com a qualidade no serviço e no atendimento ao cliente e até mesmo, ausência de personalização de ofertas.

### 1.3 Objetivo do Estudo:
Considero assim, uma oportunidade de estudo e análise exploratória dos dados de identificar padrões relacionados aos clientes que efetuaram cancelamento dos serviços, bem como, buscar compreender possíveis razões que os levaram a tais decisões.

Diante dos insights e padrões encontrados, o projeto viabilizará gerar oportunidades de ofertar soluções que mais se adequem aos clientes, tal qual, buscar ações preventivas para evitar ou reduzir o índice de churn.

### 1.4 Detalhes sobre a Base de Dados:

- Cada linha representa um cliente, cada coluna contém os atributos do cliente descritos na coluna Metadados
- Os dados brutos contêm 7.043 linhas (clientes) e 21 colunas (variáveis explicativas)
- A coluna “Churn” é o alvo de estudo.
- Os dados de entrada que complementam o estudo sobre a variável alvo, possuem características relacionadas a: identificação do cliente, dados e informações sociais, dados sobre o tipo e tempo de contrato, serviços contratados e valores (mensalidade e total pago no período do contrato)

## 3. Tratamento dos Dados

### 3.1 Renomeando Colunas

- Para facilitar a interpretação, optarei por traduzir os nomes das colunas para Português(BR)

```python
df = df.rename(columns={'customerID':'ID_cliente', 'gender':'genero', 'SeniorCitizen': 'idoso', 'Partner': 'casado_a',
                        'Dependents': 'dependentes', 'tenure': 'tempo_contrato', 'PhoneService': 'servico_telefone',
                        'MultipleLines': 'linhas_multiplas', 'InternetService': 'tipo_servico_internet', 
                        'OnlineSecurity': 'seguranca_online', 'OnlineBackup': 'backup', 
                        'DeviceProtection': 'protecao_dispositivo', 'TechSupport':'servico_suporte', 
                        'StreamingTV': 'servico_streaming_tv', 'StreamingMovies': 'filmes_streaming', 
                        'Contract': 'tipo_contrato', 'PaperlessBilling': 'fatura_digital', 'PaymentMethod': 'metodo_pagamento',
                        'MonthlyCharges': 'valor_mensalidade', 'TotalCharges': 'valor_total_contrato'})
df.head(1)
```

```python
df.columns
```

### 3.2 Verificando o tipo dos dados (Conversão dos Dados)

```python
df.dtypes
```

```python
colunas = ['genero', 'idoso', 'casado_a', 'dependentes', 'servico_telefone', 'linhas_multiplas', 'tipo_servico_internet', 
           'seguranca_online', 'backup', 'protecao_dispositivo', 'servico_suporte', 'servico_streaming_tv', 
           'filmes_streaming', 'tipo_contrato', 'fatura_digital', 'metodo_pagamento', 'Churn']
df[colunas] = df[colunas].astype('category')
df['valor_mensalidade'] = df['valor_mensalidade'].astype('float64')
```

**A coluna *"valor_total_contrato"* apresentou erro na conversão (por ter valors de células como string), portanto, precisei realizar alguns procedimentos**

```python
# Validar a existência de nulos
df['valor_total_contrato'].isnull().sum()
```

```python
# Forçar a conversão e gerar nulos nos valores que retornarem erros

df['valor_total_contrato'] = pd.to_numeric(df['valor_total_contrato'], errors='coerce')
df['valor_total_contrato'] = df['valor_total_contrato'].astype('float64')
```

```python
df.dtypes
```

### 3.3 Validar a existência de nulos, após o tratamento

```python
df.isnull().sum()
```

```python
# Como são apenas 11 linhas, optei por retirá-los da base de dados
df.dropna(subset=['valor_total_contrato'], inplace=True)
df[df['valor_total_contrato'].isna()]
```

```python
df.isnull().sum()
```

### 4.2 Análise das variáveis categóricas

```python
df.select_dtypes(include=['category']).columns
```

```python
categoricas = df.select_dtypes(include=['category']).columns
for col in categoricas:
    count_values = df[col].value_counts()
    percentual = count_values / len(df) * 100
    df_resultado = pd.DataFrame({'Valor': count_values.index, 'Contagem': count_values.values, 'Percentual': percentual.values})
    df_resultado = df_resultado.sort_values(by='Percentual', ascending=False)
    display(df_resultado)
```

### 4.3 Análises Gráficas

```python
df.columns
```

```python
colunas_grafico = ['genero', 'idoso', 'casado_a', 'dependentes', 'servico_telefone', 'linhas_multiplas',
                   'tipo_servico_internet', 'seguranca_online', 'backup', 'protecao_dispositivo', 'servico_suporte', 
                   'servico_streaming_tv', 'filmes_streaming', 'tipo_contrato', 'fatura_digital', 'metodo_pagamento']
```

```python
for col in df[colunas_grafico]:
    plt.figure(figsize=(12, 5))
    sns.countplot(x=col, hue='Churn', data=df, width=0.2)
    plt.title(f'Distribuição de Churn por {col}')
    plt.show()
```
